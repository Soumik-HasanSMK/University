#1. Handling Missing Values – Drop Rows

Definition: Remove rows that have missing or null values.

Why: Sometimes missing data is too extensive; dropping rows keeps data clean.

```python
print("Dataset Shape before using dropna() : ", df.shape)

# Drop rows with any NaN
df = df.dropna()

print("Dataset Shape after using dropna() : ", df.shape)
print("\nThe Dataset afer using dropna(): \n")
df
```

#2. Handling Missing Values – Fill with Mean

Definition: Replace numeric NaNs with the mean of the column.

Why: Keeps all rows; avoids losing data.

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df[['Patient_ID','Age']])
```

#3. Fill Missing Numeric Values with Median

Definition: Replace NaN with column median.

Why: Reduces influence of outliers.

```python
df['Weight_kg'] = pd.to_numeric(df['Weight_kg'], errors='coerce')
df['Weight_kg'] = df['Weight_kg'].fillna(df['Weight_kg'].median())
print(df[['Patient_ID','Weight_kg']])
```

#4. Fill Missing Categorical Values with Mode

Definition: Replace missing categories with the most frequent value.

Why: Maintains consistency in categorical data.

```python
df['Gender'] = df['Gender'].replace(['?', 'NULL'], pd.NA)
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
print(df[['Patient_ID','Gender']])
```

#6. Convert Data Types

Definition: Ensure columns are proper numeric, datetime, or string.

Why: Enables correct calculations and operations.

```python
df['Date_of_Visit'] = pd.to_datetime(df['Date_of_Visit'], errors='coerce')
df['Height_cm'] = pd.to_numeric(df['Height_cm'], errors='coerce')
print(df[['Patient_ID','Date_of_Visit','Height_cm']])
```

#7. Standardize Text Case

Definition: Convert text columns to consistent case (lower/upper).

Why: Helps with matching, grouping, or filtering.

```python
df['City'] = df['City'].str.strip().str.title()
print(df[['Patient_ID','City']])
```

#8. Remove Extra Spaces

Definition: Strip leading/trailing spaces in strings.

Why: Prevents mismatches in categorical analysis.

```python
df['Name'] = df['Name'].str.strip()
print(df[['Patient_ID','Name']])
```

#9. Replace Invalid Values

Definition: Replace incorrect or placeholder values (like 0, ?, bad_data) with NaN or proper values.

Why: Ensures data is valid for analysis.

```python
df['Glucose_Level'] = pd.to_numeric(df['Glucose_Level'], errors='coerce')
print(df[['Patient_ID','Glucose_Level']])
```

#10. Replace NaN and 0 with Mean

Definition: Replace NaN and 0 values with the mean of the column.

Why: To handle missing values and invalid zero entries in numeric columns.

```python
# Replace 0 with NaN first, then fill NaN with mean (excluding 0s)
df['Glucose_Level'] = df['Glucose_Level'].replace(0, np.nan)
df['Glucose_Level'] = df['Glucose_Level'].fillna(df['Glucose_Level'].mean())
print(df[['Patient_ID','Glucose_Level']])
```

#11. Handle Outliers (Cap/Clip)

Definition: Limit extreme numeric values to a reasonable range.

Why: Prevents skewed analysis.

```python
df['Age'] = df['Age'].clip(lower=1, upper=100)
print(df[['Patient_ID','Age']])
```

Why: Reduces clutter, improves efficiency.

```python
df = df.drop(columns=['Remarks'])
```

#14. Replace Text Abbreviations

Definition: Standardize abbreviations (e.g., 'Fmale' → 'Female').

Why: Avoids inconsistency.

```python
df['Gender'] = df['Gender'].replace({'Fmale':'Female','female':'Female','F':'Female','M':'Male', 'f':'Female'})
print(df[['Patient_ID','Gender']])
```

#15. Encode Categorical Data

Definition: Convert categorical variables into a numerical format that can be used by machine learning algorithms.

Why: Many machine learning models require numerical input.

```python
df['Gender_Code'] = df['Gender'].map({'Male':0,'Female':1})
print(df[['Patient_ID','Gender','Gender_Code']])
```

#16. Scale Numeric Data

Definition: Standardize or normalize numeric features to have a similar range of values.

Why: Prevents features with larger values from dominating the learning process in certain algorithms.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['Age_Scaled'] = scaler.fit_transform(df[['Age']])
print(df[['Patient_ID','Age','Age_Scaled']])
```

#17. Impute Missing Dates

Definition: Replace missing date values with a suitable value.

Why: To handle missing data in date columns, which is important for time-series analysis or other date-based operations.

```python
# Fill missing dates with the mode
df['Date_of_Visit'] = df['Date_of_Visit'].fillna(df['Date_of_Visit'].mode()[0])
print(df[['Patient_ID', 'Date_of_Visit']])
```

#18. Extract Date Features

Definition: Create new features from a datetime column (e.g., year, month, day, day of week).

Why: To capture temporal patterns and use them in analysis or modeling.

```python
df['Visit_Year'] = df['Date_of_Visit'].dt.year
df['Visit_Month'] = df['Date_of_Visit'].dt.month
df['Visit_Day'] = df['Date_of_Visit'].dt.day
print(df[['Patient_ID', 'Date_of_Visit', 'Visit_Year', 'Visit_Month', 'Visit_Day']])
```

#19. Replace Zeros in Numeric Fields

Definition: Replace zero values in numeric columns with a suitable value (e.g., mean, median, or NaN).

Why: Zero values can sometimes represent missing or invalid data, and replacing them can improve the quality of analysis.

```python
# Replace 0 with NaN in Height and Weight columns
df['Height'] = df['Height'].replace(0, np.nan)
df['Weight'] = df['Weight'].replace(0, np.nan)

# Fill NaN values with the mean (excluding the replaced zeros)
df['Height'] = df['Height'].fillna(df['Height'].mean())
df['Weight'] = df['Weight'].fillna(df['Weight'].mean())

print(df[['Patient_ID', 'Height', 'Weight']])
```

#20. Standardize Measurement Units

Definition: Remove units from numeric columns (e.g., 170cm, 65kg).

Why: Essential for accurate comparisons and calculations.

```python
df['Height'] = df['Height'].astype(str).str.replace('cm','')
df['Height'] = pd.to_numeric(df['Height'], errors='coerce')
df['Weight'] = df['Weight'].astype(str).str.replace('kg','')
df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')

print(df[['Patient_ID', 'Height', 'Weight']])
```

#21. Remove Special Characters

Definition: Remove unwanted characters in string columns.

Why: Ensures clean text for analysis.

```python
df['Name'] = df['Name'].str.replace('[^a-zA-Z\s.]','', regex=True)

print(df[['Patient_ID', 'Name']])
```

#22. Replace Nulls with Logical Values

Definition: Replace missing values with 0 or False for specific features.

Why: Some models need no NaNs.

```python
df['Blood_Pressure'] = df['Blood_Pressure'].replace('0/0', np.nan)
df['Blood_Pressure'] = df['Blood_Pressure'].fillna(df['Blood_Pressure'].mode()[0])
print(df[['Patient_ID', 'Blood_Pressure']])
```

#23. Handle Outliers Using IQR

Definition: Identify and potentially remove or cap extreme values based on the Interquartile Range (IQR).

Why: IQR is robust to outliers and provides a data-driven approach to defining the range of typical values.

```python
Q1 = df['Glucose_Level'].quantile(0.25)
Q3 = df['Glucose_Level'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['Glucose_Level'] < (Q1 - 1.5*IQR)) | (df['Glucose_Level'] > (Q3 + 1.5*IQR)))]
print(df[['Patient_ID', 'Glucose_Level']])
```

#24. Apply Transformations

Definition: Log or sqrt transformations to normalize skewed data.

Why: Helps modeling and statistical analysis.

```python
df['Log_Glucose'] = np.log1p(df['Glucose_Level'])
df['Sqrt_Glucose'] = np.sqrt(df['Glucose_Level'])
```
